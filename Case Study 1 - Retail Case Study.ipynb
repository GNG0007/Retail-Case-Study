# ##3. Generate histograms for all continuous variables and frequency bars for categorical variables

```python
Customer_Final.keys()
```

```python
num_df=Customer_Final.select_dtypes(exclude='object')
obj_num=Customer_Final.select_dtypes(include='object')
```

```python
num_df.head(1)
```

```python
obj_num.head(1)
```

```python
def continuous_variables_1(var):
            hist=sns.histplot(var)
            plt.title("histogram for " + var.name )
            plt.xticks(rotation=45)
            plt.show()
            return hist
```

```python
num_df.apply(continuous_variables_1)
plt.show()
```

```python
def categorical_variables_11(var):
            var_counts = var.value_counts() 
            var_counts.plot(kind='bar')
            plt.title("frequency bars " + var.name )
            plt.xticks(rotation=45)
            plt.show()
            return var
          
```

```python
obj_num.apply(categorical_variables_11)
plt.show()
```
